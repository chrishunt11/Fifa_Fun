## Fixing Inconsistencies

```python
df['Wage'] = df['Wage'].str.replace('€', '').str.replace('K', '').astype(float)
df['Wage'] = df['Wage'].copy().sort_values(ascending=False)
```

```python
df['Weight'] = df['Weight'].str.replace('lbs', '').astype(float)
```

```python
import numpy as np

# Function to convert feet to inches
def feet_to_inches(height):
    if pd.notna(height):
        if "'" in height:  # Check if the string contains "'"
            feet, inches = map(int, height.split("'"))
            total_inches = feet * 12 + inches
            return total_inches
        elif isinstance(height, (int, float)):  # Check if it's a numeric value
            return np.nan  # Return NaN for numeric values
        else:
            return np.nan  # Return NaN for unexpected cases
    else:
        return np.nan  # Return NaN if it's already NaN

# Convert 'Height' column to inches
df['Height'] = df['Height'].apply(feet_to_inches)
```

### Highest/Lowest Team Overall

```python
# Calculate average overall ratings for each club
club_ratings = df.groupby('Club')['Overall'].mean()

# Identify the club with the highest and lowest average overall ratings
best_club = club_ratings.idxmax()
best_rating = club_ratings.max()

worst_club = club_ratings.idxmin()
worst_rating = club_ratings.min()

# Print the results
print(f"The club with the highest average overall rating is {best_club} with a rating of {best_rating:.2f}.")
print(f"On the other hand, the club with the lowest average overall rating is {worst_club} with a rating of {worst_rating:.2f}.")
```

```python
import matplotlib.pyplot as plt

# Make sure 'Overall' column is of numeric type
df['Overall'] = pd.to_numeric(df['Overall'], errors='coerce')

# Selecting the top 10 clubs
top_10_clubs = club_ratings.nlargest(10)

# Reset the index for plotting
top_10_clubs = top_10_clubs.reset_index()

# Example color palette
colors = ['skyblue' if club != best_club else 'orange' for club in top_10_clubs['Club']]

# Plotting the bar chart for the top 10 clubs
plt.figure(figsize=(12, 8), dpi=100)
plt.bar(top_10_clubs['Club'], top_10_clubs['Overall'], color=colors, label='Highest Average Overall Rating')
plt.title('Top 10 Clubs by Average Overall Ratings', fontsize=16)
plt.xlabel('Club', fontsize=12)
plt.ylabel('Average Overall Rating', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.annotate(f'Highest: {best_club} ({best_rating:.2f})', 
             xy=(0, best_rating), xytext=(0, best_rating + 5),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=12, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper right')
plt.tight_layout()

plt.show()
```

### Distribution of Age/Height/Weight

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort the DataFrame by Height
df_sorted_height = df.sort_values('Height', ascending=True)

# Plotting the distribution of Age
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')

# Plotting the distribution of Height
plt.subplot(2, 2, 2)
sns.histplot(df_sorted_height['Height'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Height')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Sort the DataFrame by Weight
df_sorted_weight = df.sort_values('Weight', ascending=False)

# Plotting the distribution of Weight using Seaborn histplot
plt.subplot(2, 2, 3)
sns.histplot(df_sorted_weight['Weight'], bins=range(100, 275, 5), kde=True, color='lightgreen', edgecolor='black')
plt.title('Distribution of Weight')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better aesthetics
sns.set(style="whitegrid", font_scale=1.2)

# Scatter plot with regression line
plt.figure(figsize=(12, 8))
sns.regplot(x='Skill Moves', y='Overall', data=df, scatter_kws={'s': 50, 'alpha': 0.7, 'color': 'skyblue'}, 
            line_kws={'color': 'red', 'linewidth': 2})
plt.title('Relationship Between Skill Moves and Overall Rating', fontsize=16, fontweight='bold')
plt.xlabel('Skill Moves', fontsize=14)
plt.ylabel('Overall Rating', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='both', linestyle='--', alpha=0.5)

# Add a subtle background color
plt.axhline(0, color='white', linewidth=2, linestyle='--')

# Tighten layout for better presentation
plt.tight_layout()

# Show the plot
plt.show()
```
